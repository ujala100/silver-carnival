```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
iris = load_iris()
X=iris.data
y=iris.target
print(iris)
sc=StandardScaler()
X_std=sc.fit_transform(X)
print(X_std)
```

```python
from sklearn.cluster import AgglomerativeClustering
agg=AgglomerativeClustering(n_clusters=3,linkage='ward')
clusters=agg.fit_predict(X_std)
print(clusters)
```

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram,linkage
from sklearn.decomposition import PCA
plt.figure(figsize=(12,5))
linkage_matrix=linkage(X_std,method='ward')
dendrogram(linkage_matrix,truncate_mode='level',p=5)
plt.xlabel("Simple Index")
plt.ylabel("Distance")
plt.show()
```

```python
pca=PCA(n_components=2)
X_pca=pca.fit_transform(X_std)
plt.figure(figsize=(8,5))
plt.scatter(X_pca[:,0],X_pca[:,1],c=clusters,cmap='viridis',edgecolor='k')
plt.title("2D Cluster Visualization (PCA-reduced)")
plt.xlabel(" Principal Component 1")
plt.ylabel(" Principal Component 2")
plt.show()
```
